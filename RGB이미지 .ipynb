# 1. RGB 이미지

RGB 이미지는 빨강(Red), 초록(Green), 파랑(Blue) 세 가지 기본 색상 채널로 구성된 디지털 이미지입니다. 

이 세 가지 색상의 조합을 통해 모든 색상을 표현할 수 있습니다. 

RGB는 가산 색 혼합 방식에 기반하며, 각 채널의 값이 0에서 255 사이의 정수 값으로 표현될 때 가장 일반적으로 사용됩니다.

### 1) 3X2 픽셀

코드설명

```python
# RGB 이미지 데이터 생성 (3x2 픽셀)
# rgb_image = np.array([...], dtype=np.uint8)
# RGB 이미지 데이터 생성:
# np.array를 사용해 2차원 배열을 생성합니다. 배열의 각 요소는 [R, G, B] 값을 가지는 픽셀입니다.
# dtype=np.uint8로 지정하여 배열의 데이터 타입을 8비트 정수 (uint8)로 설정합니다. 
# 이는 일반적인 이미지 데이터 표현 방식이며, 각 채널 값은 0~255 범위를 가집니다.
```

```python
# 이미지를 화면에 표시
# plt.imshow(rgb_image)
# 이미지 표시: plt.imshow(rgb_image)는 rgb_image 배열을 이미지로 화면에 표시합니다. 
# 배열의 각 (R, G, B) 값이 해당 색상 픽셀로 렌더링됩니다.

# # plt.axis('off')
# 축 제거: plt.axis('off')는 이미지를 표시할 때 x축과 y축을 표시하지 않도록 설정합니다. 
# 이렇게 하면 이미지만 깔끔하게 화면에 보이게 됩니다.
    
# plt.title("Example RGB Image (3x2 Pixels)")
# 제목 추가: plt.title()을 사용하여 이미지 위에 제목을 추가합니다. 
# 여기서는 "Example RGB Image (3x2 Pixels)"이라는 제목을 추가하여 이미지가 무엇인지 설명합니다.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# RGB 이미지 데이터 생성 (3x2 픽셀)
rgb_image = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],  # 첫 번째 행: 빨강, 초록, 파랑
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]]  # 두 번째 행: 노랑, 청록, 자홍
], dtype=np.uint8)

# 이미지를 화면에 표시
plt.imshow(rgb_image)
plt.axis('off')  # 축 제거
plt.title("Example RGB Image (3x2 Pixels)")
plt.show()
```

### 2) 3X3 하얀색 픽셀

```python
import numpy as np
import matplotlib.pyplot as plt

# 흰색 이미지 데이터 생성 (3x3 픽셀)
white_image = np.full((3, 3, 3), 255, dtype=np.uint8)  # 모든 채널 값을 255로 설정

# 흰색 이미지를 화면에 표시
plt.imshow(white_image)
plt.axis('off')  # 축 제거
plt.title("White Image (3x3 Pixels)")
plt.show()
```

# 2. 그레이스케일(흑백) 이미지

RGB와 같은 다중 색상 채널을 사용하지 않고, 각 픽셀을 하나의 밝기 값으로 표현하는 이미지입니다. 

그레이스케일 이미지는 흑백 이미지라고도 하며, 주로 흰색과 검정색 사이의 다양한 회색 음영을 통해 이미지를 구성합니다.

각 픽셀 값은 0에서 255 사이의 단일 값으로, 명암을 나타냅니다. 

값이 0이면 완전한 검정색, 값이 255이면 완전한 흰색을 나타내며, 그 사이의 값들은 점진적인 회색 음영을 나타냅니다.

그레이스케일 이미지는 색상이 아니라 밝기 정보만을 포함합니다.

예를 들어, 값이 128인 픽셀은 중간 밝기의 회색을 나타내며, 0은 검정색, 255는 흰색을 의미합니다.

데이터 크기:
RGB 이미지와 달리, 그레이스케일 이미지는 단일 값으로 표현되므로 데이터 크기가 작습니다.

예를 들어, 같은 크기의 RGB 이미지는 3개의 값을 저장해야 하지만, 그레이스케일 이미지는 한 값만 저장하면 됩니다. 이는 메모리와 연산 비용 면에서 효율적입니다.

RGB와 그레이스케일 이미지의 차이

RGB 이미지: 세 개의 색상 채널(Red, Green, Blue)로 구성되며, 각 픽셀은 (R, G, B)로 표현됩니다. 모든 색상을 표현할 수 있어 컬러 이미지를 나타내는 데 사용됩니다.
그레이스케일 이미지: 단일 채널로 구성되어 밝기 정보만을 포함합니다. 색상이 없는 회색 음영으로만 이루어져 있습니다.

```python
import numpy as np
import matplotlib.pyplot as plt

# RGB 이미지 데이터 생성 (3x3 픽셀)
rgb_image = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],    # 빨강, 초록, 파랑
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]],  # 노랑, 청록, 자홍
    [[128, 128, 128], [64, 64, 64], [0, 0, 0]]   # 회색, 어두운 회색, 검정
], dtype=np.uint8)

# 칼라 이미지 시각화
plt.imshow(rgb_image)
plt.axis('off')
plt.title("Color Image (3x3 Pixels)")
plt.show()

# 그레이 이미지 데이터 생성 (3x3 픽셀)
gray_image = np.array([
    [50, 150, 200],   # 첫 번째 행
    [100, 180, 220],  # 두 번째 행
    [30, 120, 250]    # 세 번째 행
], dtype=np.uint8)

# 그레이 이미지를 화면에 표시 (cmap='gray' 사용)
plt.imshow(gray_image, cmap='gray')
plt.axis('off')  # 축 제거
plt.title("Grayscale Image (3x3 Pixels)")
plt.show()
```

# 3. RGB 이미지를 그레이스케일로 변환

RGB 이미지를 그레이스케일로 변환하는 과정은 각 픽셀의 RGB 값을 단일 밝기 값으로 바꾸는 것을 의미합니다. 

이 과정에서 주로 사용하는 방식은 **가중 합(weighted sum)**을 사용하는 방법입니다. 

사람이 각 색상에 대해 느끼는 밝기의 차이를 고려하여 가중치를 부여합니다.

변환 공식
RGB를 그레이스케일로 변환할 때 가장 널리 사용되는 공식은 다음과 같습니다:

Grayscale Value=0.2989×R +0.5870×𝐺+0.1140×𝐵

R, G, B: 빨강, 초록, 파랑의 각각의 값 (0~255 사이).
각 색상의 가중치는 사람의 눈이 색을 느끼는 민감도를 반영합니다.
예를 들어, 녹색에 대한 감도가 가장 높기 때문에 초록(G)에 가장 높은 가중치 0.587이 사용됩니다.

### 2) 수동으로 가중치를 적용하여 변환

```python
import numpy as np
import matplotlib.pyplot as plt

# RGB 이미지 데이터 생성 (3x3 픽셀)
rgb_image = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],    # 빨강, 초록, 파랑
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]],  # 노랑, 청록, 자홍
    [[128, 128, 128], [64, 64, 64], [0, 0, 0]]   # 회색, 어두운 회색, 검정
], dtype=np.uint8)

# 칼라 이미지 시각화
plt.imshow(rgb_image)
plt.axis('off')
plt.title("Color Image (3x3 Pixels)")
plt.show()

# 그레이스케일로 변환 (가중치 적용)
gray_image = np.dot(rgb_image[..., :3], [0.2989, 0.5870, 0.1140]).astype(np.uint8)

# 그레이스케일 이미지 시각화
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image (Converted from Color)")
plt.show()
```
