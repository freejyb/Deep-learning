# 2. Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº(ÌùëÎ∞±) Ïù¥ÎØ∏ÏßÄ

RGBÏôÄ Í∞ôÏùÄ Îã§Ï§ë ÏÉâÏÉÅ Ï±ÑÎÑêÏùÑ ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÍ≥†, Í∞Å ÌîΩÏÖÄÏùÑ ÌïòÎÇòÏùò Î∞ùÍ∏∞ Í∞íÏúºÎ°ú ÌëúÌòÑÌïòÎäî Ïù¥ÎØ∏ÏßÄÏûÖÎãàÎã§. 

Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Ïù¥ÎØ∏ÏßÄÎäî ÌùëÎ∞± Ïù¥ÎØ∏ÏßÄÎùºÍ≥†ÎèÑ ÌïòÎ©∞, Ï£ºÎ°ú Ìù∞ÏÉâÍ≥º Í≤ÄÏ†ïÏÉâ ÏÇ¨Ïù¥Ïùò Îã§ÏñëÌïú ÌöåÏÉâ ÏùåÏòÅÏùÑ ÌÜµÌï¥ Ïù¥ÎØ∏ÏßÄÎ•º Íµ¨ÏÑ±Ìï©ÎãàÎã§.

Í∞Å ÌîΩÏÖÄ Í∞íÏùÄ 0ÏóêÏÑú 255 ÏÇ¨Ïù¥Ïùò Îã®Ïùº Í∞íÏúºÎ°ú, Î™ÖÏïîÏùÑ ÎÇòÌÉÄÎÉÖÎãàÎã§. 

Í∞íÏù¥ 0Ïù¥Î©¥ ÏôÑÏ†ÑÌïú Í≤ÄÏ†ïÏÉâ, Í∞íÏù¥ 255Ïù¥Î©¥ ÏôÑÏ†ÑÌïú Ìù∞ÏÉâÏùÑ ÎÇòÌÉÄÎÇ¥Î©∞, Í∑∏ ÏÇ¨Ïù¥Ïùò Í∞íÎì§ÏùÄ Ï†êÏßÑÏ†ÅÏù∏ ÌöåÏÉâ ÏùåÏòÅÏùÑ ÎÇòÌÉÄÎÉÖÎãàÎã§.

Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Ïù¥ÎØ∏ÏßÄÎäî ÏÉâÏÉÅÏù¥ ÏïÑÎãàÎùº Î∞ùÍ∏∞ Ï†ïÎ≥¥ÎßåÏùÑ Ìè¨Ìï®Ìï©ÎãàÎã§.

ÏòàÎ•º Îì§Ïñ¥, Í∞íÏù¥ 128Ïù∏ ÌîΩÏÖÄÏùÄ Ï§ëÍ∞Ñ Î∞ùÍ∏∞Ïùò ÌöåÏÉâÏùÑ ÎÇòÌÉÄÎÇ¥Î©∞, 0ÏùÄ Í≤ÄÏ†ïÏÉâ, 255Îäî Ìù∞ÏÉâÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.

Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞:
RGB Ïù¥ÎØ∏ÏßÄÏôÄ Îã¨Î¶¨, Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Ïù¥ÎØ∏ÏßÄÎäî Îã®Ïùº Í∞íÏúºÎ°ú ÌëúÌòÑÎêòÎØÄÎ°ú Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞Í∞Ä ÏûëÏäµÎãàÎã§.

ÏòàÎ•º Îì§Ïñ¥, Í∞ôÏùÄ ÌÅ¨Í∏∞Ïùò RGB Ïù¥ÎØ∏ÏßÄÎäî 3Í∞úÏùò Í∞íÏùÑ Ï†ÄÏû•Ìï¥Ïïº ÌïòÏßÄÎßå, Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Ïù¥ÎØ∏ÏßÄÎäî Ìïú Í∞íÎßå Ï†ÄÏû•ÌïòÎ©¥ Îê©ÎãàÎã§. Ïù¥Îäî Î©îÎ™®Î¶¨ÏôÄ Ïó∞ÏÇ∞ ÎπÑÏö© Î©¥ÏóêÏÑú Ìö®Ïú®Ï†ÅÏûÖÎãàÎã§.

RGBÏôÄ Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Ïù¥ÎØ∏ÏßÄÏùò Ï∞®Ïù¥

RGB Ïù¥ÎØ∏ÏßÄ: ÏÑ∏ Í∞úÏùò ÏÉâÏÉÅ Ï±ÑÎÑê(Red, Green, Blue)Î°ú Íµ¨ÏÑ±ÎêòÎ©∞, Í∞Å ÌîΩÏÖÄÏùÄ (R, G, B)Î°ú ÌëúÌòÑÎê©ÎãàÎã§. Î™®Îì† ÏÉâÏÉÅÏùÑ ÌëúÌòÑÌï† Ïàò ÏûàÏñ¥ Ïª¨Îü¨ Ïù¥ÎØ∏ÏßÄÎ•º ÎÇòÌÉÄÎÇ¥Îäî Îç∞ ÏÇ¨Ïö©Îê©ÎãàÎã§.
Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Ïù¥ÎØ∏ÏßÄ: Îã®Ïùº Ï±ÑÎÑêÎ°ú Íµ¨ÏÑ±ÎêòÏñ¥ Î∞ùÍ∏∞ Ï†ïÎ≥¥ÎßåÏùÑ Ìè¨Ìï®Ìï©ÎãàÎã§. ÏÉâÏÉÅÏù¥ ÏóÜÎäî ÌöåÏÉâ ÏùåÏòÅÏúºÎ°úÎßå Ïù¥Î£®Ïñ¥Ï†∏ ÏûàÏäµÎãàÎã§.

```python
import numpy as np
import matplotlib.pyplot as plt

# RGB Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (3x3 ÌîΩÏÖÄ)
rgb_image = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],    # Îπ®Í∞ï, Ï¥àÎ°ù, ÌååÎûë
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]],  # ÎÖ∏Îûë, Ï≤≠Î°ù, ÏûêÌôç
    [[128, 128, 128], [64, 64, 64], [0, 0, 0]]   # ÌöåÏÉâ, Ïñ¥ÎëêÏö¥ ÌöåÏÉâ, Í≤ÄÏ†ï
], dtype=np.uint8)

# ÏπºÎùº Ïù¥ÎØ∏ÏßÄ ÏãúÍ∞ÅÌôî
plt.imshow(rgb_image)
plt.axis('off')
plt.title("Color Image (3x3 Pixels)")
plt.show()

# Í∑∏Î†àÏù¥ Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (3x3 ÌîΩÏÖÄ)
gray_image = np.array([
    [50, 150, 200],   # Ï≤´ Î≤àÏß∏ Ìñâ
    [100, 180, 220],  # Îëê Î≤àÏß∏ Ìñâ
    [30, 120, 250]    # ÏÑ∏ Î≤àÏß∏ Ìñâ
], dtype=np.uint8)

# Í∑∏Î†àÏù¥ Ïù¥ÎØ∏ÏßÄÎ•º ÌôîÎ©¥Ïóê ÌëúÏãú (cmap='gray' ÏÇ¨Ïö©)
plt.imshow(gray_image, cmap='gray')
plt.axis('off')  # Ï∂ï Ï†úÍ±∞
plt.title("Grayscale Image (3x3 Pixels)")
plt.show()
```

# 3. RGB Ïù¥ÎØ∏ÏßÄÎ•º Í∑∏Î†àÏù¥Ïä§ÏºÄÏùºÎ°ú Î≥ÄÌôò

RGB Ïù¥ÎØ∏ÏßÄÎ•º Í∑∏Î†àÏù¥Ïä§ÏºÄÏùºÎ°ú Î≥ÄÌôòÌïòÎäî Í≥ºÏ†ïÏùÄ Í∞Å ÌîΩÏÖÄÏùò RGB Í∞íÏùÑ Îã®Ïùº Î∞ùÍ∏∞ Í∞íÏúºÎ°ú Î∞îÍæ∏Îäî Í≤ÉÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§. 

Ïù¥ Í≥ºÏ†ïÏóêÏÑú Ï£ºÎ°ú ÏÇ¨Ïö©ÌïòÎäî Î∞©ÏãùÏùÄ **Í∞ÄÏ§ë Ìï©(weighted sum)**ÏùÑ ÏÇ¨Ïö©ÌïòÎäî Î∞©Î≤ïÏûÖÎãàÎã§. 

ÏÇ¨ÎûåÏù¥ Í∞Å ÏÉâÏÉÅÏóê ÎåÄÌï¥ ÎäêÎÅºÎäî Î∞ùÍ∏∞Ïùò Ï∞®Ïù¥Î•º Í≥†Î†§ÌïòÏó¨ Í∞ÄÏ§ëÏπòÎ•º Î∂ÄÏó¨Ìï©ÎãàÎã§.

Î≥ÄÌôò Í≥µÏãù
RGBÎ•º Í∑∏Î†àÏù¥Ïä§ÏºÄÏùºÎ°ú Î≥ÄÌôòÌï† Îïå Í∞ÄÏû• ÎÑêÎ¶¨ ÏÇ¨Ïö©ÎêòÎäî Í≥µÏãùÏùÄ Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§:

Grayscale¬†Value=0.2989√óR +0.5870√óùê∫+0.1140√óùêµ

R, G, B: Îπ®Í∞ï, Ï¥àÎ°ù, ÌååÎûëÏùò Í∞ÅÍ∞ÅÏùò Í∞í (0~255 ÏÇ¨Ïù¥).
Í∞Å ÏÉâÏÉÅÏùò Í∞ÄÏ§ëÏπòÎäî ÏÇ¨ÎûåÏùò ÎààÏù¥ ÏÉâÏùÑ ÎäêÎÅºÎäî ÎØºÍ∞êÎèÑÎ•º Î∞òÏòÅÌï©ÎãàÎã§.
ÏòàÎ•º Îì§Ïñ¥, ÎÖπÏÉâÏóê ÎåÄÌïú Í∞êÎèÑÍ∞Ä Í∞ÄÏû• ÎÜíÍ∏∞ ÎïåÎ¨∏Ïóê Ï¥àÎ°ù(G)Ïóê Í∞ÄÏû• ÎÜíÏùÄ Í∞ÄÏ§ëÏπò 0.587Ïù¥ ÏÇ¨Ïö©Îê©ÎãàÎã§.

### 2) ÏàòÎèôÏúºÎ°ú Í∞ÄÏ§ëÏπòÎ•º Ï†ÅÏö©ÌïòÏó¨ Î≥ÄÌôò

```python
import numpy as np
import matplotlib.pyplot as plt

# RGB Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (3x3 ÌîΩÏÖÄ)
rgb_image = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],    # Îπ®Í∞ï, Ï¥àÎ°ù, ÌååÎûë
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]],  # ÎÖ∏Îûë, Ï≤≠Î°ù, ÏûêÌôç
    [[128, 128, 128], [64, 64, 64], [0, 0, 0]]   # ÌöåÏÉâ, Ïñ¥ÎëêÏö¥ ÌöåÏÉâ, Í≤ÄÏ†ï
], dtype=np.uint8)

# ÏπºÎùº Ïù¥ÎØ∏ÏßÄ ÏãúÍ∞ÅÌôî
plt.imshow(rgb_image)
plt.axis('off')
plt.title("Color Image (3x3 Pixels)")
plt.show()

# Í∑∏Î†àÏù¥Ïä§ÏºÄÏùºÎ°ú Î≥ÄÌôò (Í∞ÄÏ§ëÏπò Ï†ÅÏö©)
gray_image = np.dot(rgb_image[..., :3], [0.2989, 0.5870, 0.1140]).astype(np.uint8)

# Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Ïù¥ÎØ∏ÏßÄ ÏãúÍ∞ÅÌôî
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image (Converted from Color)")
plt.show()
```
