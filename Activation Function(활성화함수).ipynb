# 활성화 함수 정의

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    e_x = np.exp(x - np.max(x))  # For numerical stability
    return e_x / e_x.sum(axis=0)
```

Leaky ReLU 함수에서 alpha

```python
# Leaky ReLU 함수에서 alpha는 기울기 계수로, 입력 값이 음수일 때 적용되는 작은 기울기를 나타냅니다.
# alpha의 의미:
# alpha는 음수 입력에 대해 얼마나 큰 음수 출력을 허용할지를 결정합니다.
# 일반적으로 ReLU 함수는 음수 값을 모두 0으로 변환하는데, 이 경우 학습 중에 음수 영역의 뉴런들이 죽을 수(dead neurons) 있습니다. 이 문제를 해결하기 위해 Leaky ReLU는 음수 값에 대해서도 작은 기울기를 허용하여, 네트워크가 완전히 죽지 않도록 만듭니다.

# alpha 값이 클수록 음수 입력에 대한 기울기가 커지며, 반대로 alpha 값이 작을수록 음수 입력에 대해 출력이 거의 0에 가까워집니다.
# Leaky ReLU의 활용:
# alpha 값은 일반적으로 0.01로 설정되지만, 실험을 통해 최적의 값을 찾을 수 있습니다.
# 음수 값이 중요할 가능성이 있는 데이터나 모델 구조에서는 Leaky ReLU가 일반적인 ReLU보다 더 좋은 성능을 보일 수 있습니다.
```

# 하이퍼볼릭 탄젠트 함수 (Hyperbolic Tangent, tanh)

하이퍼볼릭 탄젠트 함수 (tanh)는 신경망에서 많이 사용되는 활성화 함수 중 하나입니다. 

출력 범위: -1에서 1

중심성: 함수의 출력 값이 0을 중심으로 대칭이므로, 입력 값이 음수일 때는 음수를 출력하고, 양수일 때는 양수를 출력합니다.
    
미분 가능: 신경망에서 역전파(backpropagation)를 수행할 때 tanh 함수는 미분이 가능하다는 장점이 있습니다.

2. 특징 및 장점

출력 값의 범위가 -1에서 1 사이이므로, 데이터가 0을 중심으로 조정됩니다. 이는 학습을 더 안정적으로 만듭니다.

Sigmoid 함수와 비교할 때, 기울기 소실 문제(vanishing gradient problem)가 덜 발생합니다.
    
주로 은닉층의 활성화 함수로 사용됩니다.
    
    

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 범위 설정
x = np.linspace(-5, 5, 100)

# tanh 함수 적용
y = np.tanh(x)

# tanh 함수의 미분 계산
y_derivative = 1 - y**2

# 그래프 그리기
plt.figure(figsize=(12, 6))

# tanh 함수 그래프
plt.subplot(1, 2, 1)
plt.plot(x, y, color='blue', label='tanh(x)')
plt.title('Hyperbolic Tangent Function (tanh)')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

# tanh 함수의 미분 그래프
plt.subplot(1, 2, 2)
plt.plot(x, y_derivative, color='red', label="tanh'(x)")
plt.title('Derivative of tanh')
plt.xlabel('x')
plt.ylabel("tanh'(x)")
plt.grid()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

그래프 시각화:
왼쪽 그래프는 tanh 함수의 그래프입니다.
    
오른쪽 그래프는 tanh 함수의 미분 그래프입니다.

결과 분석

tanh 함수 그래프: x값이 증가할수록 1에 수렴하고, 감소할수록 -1에 수렴합니다.
    
미분 그래프: 미분 값은 x = 0 근처에서 최대값(1)을 가지며, x값이 멀어질수록 0에 가까워집니다.

# ReLU 함수와 그 미분 함수

```python
# 코드 설명
# ReLU 함수 정의:
# relu(x)는 입력 값 중 0보다 큰 값은 그대로 반환하고, 0 이하의 값은 0으로 반환합니다.
# 미분 함수 정의:
# relu_derivative(x)는 입력 값이 0보다 크면 1, 그렇지 않으면 0을 반환합니다.
# 그래프 시각화:
# 왼쪽 그래프는 ReLU 함수의 출력 값을 보여줍니다.
# 오른쪽 그래프는 ReLU 함수의 미분 값을 나타냅니다.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ReLU 함수 정의
def relu(x):
    return np.maximum(0, x)

# ReLU 함수의 미분 정의
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# x 값 범위 설정
x = np.linspace(-5, 5, 100)

# ReLU 함수와 미분 계산
y = relu(x)
y_derivative = relu_derivative(x)

# 그래프 그리기
plt.figure(figsize=(12, 6))

# ReLU 함수 그래프
plt.subplot(1, 2, 1)
plt.plot(x, y, color='blue', label='ReLU(x)')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

# ReLU 함수의 미분 그래프
plt.subplot(1, 2, 2)
plt.plot(x, y_derivative, color='red', label="ReLU'(x)")
plt.title('Derivative of ReLU')
plt.xlabel('x')
plt.ylabel("ReLU'(x)")
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

# 소프트맥스(Softmax) 함수

소프트맥스 함수는 입력 값들의 집합을 확률 값으로 변환하며, 출력 값의 합이 1이 됩니다.

소프트맥스 함수는 다중 클래스 분류 문제에서 출력층의 활성화 함수로 많이 사용되며, 각 클래스에 대한 확률을 계산하는 데 유용합니다.

```python
import numpy as np
import matplotlib.pyplot as plt

# 소프트맥스 함수 정의
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # 오버플로우 방지용
    return exp_x / np.sum(exp_x, axis=0)

# x 값 범위 설정
x = np.linspace(-2, 2, 100)
X = np.vstack([x, x + 1, x - 1])  # 3개의 입력 값 집합 생성

# 소프트맥스 함수 적용
y = softmax(X)

# 그래프 그리기
plt.figure(figsize=(12, 6))

# 소프트맥스 함수 그래프
for i in range(X.shape[0]):
    plt.plot(x, y[i], label=f"Softmax Component {i + 1}")

plt.title('Softmax Function')
plt.xlabel('x')
plt.ylabel('Softmax Output')
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.show()
```

결과 분석
소프트맥스 함수는 입력 값이 클수록 해당 출력 값이 높아지고, 나머지 값의 확률은 낮아집니다.
확률 값의 합은 항상 1이 됩니다.
주로 다중 클래스 분류 문제에서, 각 클래스의 확률을 계산할 때 사용됩니다.

## sigmoid() 함수 

```python
# x_values: 예시 입력 값 배열입니다. 이 값을 사용하여 시그모이드 함수의 출력 결과를 계산합니다.
# sigmoid_values: 시그모이드 함수의 계산 결과 값입니다.
```

```python
import numpy as np

# 시그모이드 함수 정의
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 예시 입력 값
x_values = np.array([-2, -1, 0, 1, 2])

# 시그모이드 함수 적용
sigmoid_values = sigmoid(x_values)

# 결과 출력
for x, sigmoid_value in zip(x_values, sigmoid_values):
    print(f"x = {x}, sigmoid(x) = {sigmoid_value:.4f}")
```

```python
# 결과 해석:
# 입력 값이 0일 때 시그모이드 함수의 출력은 0.5입니다.
# 입력 값이 음수일수록 출력 값은 0에 가까워지고, 양수일수록 출력 값은 1에 가까워집니다.
# 이는 시그모이드 함수가 입력을 0과 1 사이의 확률 값으로 변환하는 데 사용되는 것을 보여줍니다.
```

## ReLU (Rectified Linear Unit) 함수

```python
# 코드 설명:
# ReLU 함수 정의: np.maximum(0, x)를 사용하여 
# 𝑥 x가 0보다 크면 그 값을, 그렇지 않으면 0을 반환합니다.
# 예시 입력 값: x_values 배열은 [-2, -1, 0, 1, 2, 3]과 같은 값을 가지고 있습니다.
# ReLU 함수 적용: 입력 값 배열에 ReLU 함수를 적용하여 변환된 값을 반환합니다.
```

```python
import numpy as np

# ReLU 함수 정의
def relu(x):
    return np.maximum(0, x)

# 예시 입력 값
x_values = np.array([-2, -1, 0, 1, 2, 3])

# ReLU 함수 적용
relu_values = relu(x_values)

# 결과 출력
print("입력 값:", x_values)
print("ReLU 적용 결과:", relu_values)
```

```python
# 입력 값이 음수인 경우: ReLU 함수는 음수 값을 0으로 변환합니다. 예를 들어, -2와 -1은 0으로 변환됩니다.
# 입력 값이 0인 경우: ReLU 함수는 0을 그대로 반환합니다.
# 입력 값이 양수인 경우: ReLU 함수는 양수 값을 그대로 반환합니다. 예를 들어, 1, 2, 3은 그대로 출력됩니다.
```

## Leaky ReLU (Leaky Rectified Linear Unit) 함수

```python
# 코드 설명:
# leaky_relu() 함수:
# 입력 x가 0보다 크면 그대로 반환하고, 0보다 작으면 
# 𝛼x를 반환합니다.
# 여기서 alpha는 0보다 작은 값에 적용되는 기울기입니다. 기본적으로 0.01로 설정되었습니다.
# 예시 입력 값:  [-3, -1, 0, 1, 3] 값을 예시로 사용하여 Leaky ReLU 함수를 적용합니다.
# 결과: 
# 각 입력 값에 대해 Leaky ReLU 함수가 어떻게 동작하는지 확인할 수 있습니다.
```

```python
import numpy as np

# Leaky ReLU 함수 정의
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# 예시 값으로 Leaky ReLU 함수 적용
example_values = np.array([-3, -1, 0, 1, 3])

# 함수 적용
leaky_relu_values = leaky_relu(example_values)

# 결과 출력
print("입력 값: ", example_values)
print("Leaky ReLU 출력 값: ", leaky_relu_values)
```

```python
# 설명:
# 음수 값 -3과 -1에 대해서는 alpha=0.01이 적용되어 각각 -0.03과 -0.01이 출력되었습니다.
# 0은 그대로 0으로 출력됩니다.
# 양수 값 1과 3은 그대로 출력됩니다.
```

## softmax 함수

```python
# 설명:
# logits = np.array([2.0, 1.0, 0.1]): 이 배열은 3개의 클래스에 대한 예측 값(로짓 값)을 의미합니다. 각 클래스는 2.0, 1.0, 0.1의 값을 가집니다.
# softmax(logits) 함수는 이 값을 입력으로 받아 소프트맥스 확률을 계산합니다.
```

```python
# 소프트맥스 함수 동작 원리:
# 각 로짓 값에 대해 지수 함수를 취하여 양수로 변환됩니다.
# 지수 함수로 변환된 값들을 모두 더하고, 각 지수 값을 그 합으로 나누어 확률 값을 구합니다.
# 그 결과는 각 클래스에 속할 확률을 나타내며, 이 확률들의 합은 항상 1이 됩니다.
```

```python
import numpy as np

# 소프트맥스 함수 정의
def softmax(logits):
    exps = np.exp(logits)  # 입력 값을 지수 함수로 변환
    return exps / np.sum(exps)  # 지수 값들의 합으로 나눈다

# 예시 값 (로짓 값)
logits = np.array([2.0, 1.0, 0.1])

# 소프트맥스 계산
probabilities = softmax(logits)

# 결과 출력
print("입력 로짓 값:", logits)
print("소프트맥스 확률 값:", probabilities)
print("확률 값의 합:", np.sum(probabilities))
```

```python
# 코드 실행 결과:
# 입력 로짓 값: [2.0, 1.0, 0.1]
# 클래스 1: 2.0
# 클래스 2: 1.0
# 클래스 3: 0.1

# 소프트맥스 확률 값: [0.65900114, 0.24243297, 0.09856589]
# 클래스 1의 확률: 약 65.9%
# 클래스 2의 확률: 약 24.2%
# 클래스 3의 확률: 약 9.9%

# 확률 값의 합: 1.0 (항상 1이 되어야 함)
```
