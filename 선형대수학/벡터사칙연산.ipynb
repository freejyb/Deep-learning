### 벡터 사칙연산

#### 덧셈, 뺄셈, Norm, 곱(내적), 단위벡터, 정사영(투영)

```python
# 벡터 정의
# a = np.array([4, 0])
# b = np.array([3, 4])
```

```python
# 벡터 사칙연산 + 내적 예제 (Python / NumPy)

import numpy as np

# 벡터 정의
a = np.array([4, 0])
b = np.array([3, 4])

print("벡터 a =", a)
print("벡터 b =", b)
print("-" * 40)

# 1. 벡터 덧셈
add = a + b
print("1) 벡터 덧셈 a + b =", add)
print("   의미: 두 벡터를 이어 붙인 결과\n")

# 2. 벡터 뺄셈
sub = b - a
print("2) 벡터 뺄셈 b - a =", sub)
print("   의미: a에서 b로 가는 방향 벡터\n")

# 3. 스칼라 곱
scalar = 2
scalar_mul = scalar * b
print("3) 스칼라 곱 2 * b =", scalar_mul)
print("   의미: 방향은 같고 길이는 2배\n")

# 4. 벡터의 크기 (Norm)
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)
print("4) 벡터의 크기")
print("   ||a|| =", norm_a)
print("   ||b|| =", norm_b)
print()

# 5. 내적 (dot product)
dot = np.dot(a, b)
print("5) 내적 a · b =", dot)
print("   의미: a 방향으로 b가 얼마나 향하는지\n")

# 6. 각도 계산 (내적 활용)
cos_theta = dot / (norm_a * norm_b)
theta = np.degrees(np.arccos(cos_theta))
print("6) a와 b 사이 각도")
print("   cos(theta) =", cos_theta)
print("   theta (degrees) =", theta)
print()

# 7. 단위 벡터
unit_b = b / norm_b
print("7) 단위 벡터")
print("   b의 단위벡터 =", unit_b)
print("   길이 =", np.linalg.norm(unit_b))
print()

# 8. 정사영 (투영 projection)
proj_b_on_a = (np.dot(b, a) / np.dot(a, a)) * a
print("8) b를 a에 투영 (projection)")
print("   proj_a(b) =", proj_b_on_a)
print()

# 9. 직교 성분
perp = b - proj_b_on_a
print("9) b의 a에 수직인 성분")
print("   b - proj_a(b) =", perp)
print("   a · (수직성분) =", np.dot(a, perp), "(≈ 0 → 직교)")
print()

print("✔ 요약:")
print("b = proj_a(b) + (b - proj_a(b))")
```

### 코사인 유사도 & 유클리드 유사도 예제 

```python
# 코사인 유사도 & 유클리드 유사도 예제 (Python / NumPy)

import numpy as np

# -----------------------------
# 1. 예제 벡터 정의
# -----------------------------
v1 = np.array([3, 4])
v2 = np.array([6, 8])    # v1과 방향이 같은 벡터
v3 = np.array([4, -3])   # v1과 다른 방향의 벡터

print("v1 =", v1)
print("v2 =", v2)
print("v3 =", v3)
print("-" * 50)

# -----------------------------
# 2. 코사인 유사도 (Cosine Similarity)
# -----------------------------
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

cos_v1_v2 = cosine_similarity(v1, v2)
cos_v1_v3 = cosine_similarity(v1, v3)

print("① 코사인 유사도 (Cosine Similarity)")
print("cos(v1, v2) =", cos_v1_v2)
print("→ 방향이 같음 (값 ≈ 1)")
print()
print("cos(v1, v3) =", cos_v1_v3)
print("→ 거의 직교 / 다른 방향")
print()

# -----------------------------
# 3. 유클리드 거리 (Euclidean Distance)
#    ※ 흔히 '유클리드 유사도'라고도 부름
# -----------------------------
def euclidean_distance(a, b):
    return np.linalg.norm(a - b)

dist_v1_v2 = euclidean_distance(v1, v2)
dist_v1_v3 = euclidean_distance(v1, v3)

print("② 유클리드 거리 (Euclidean Distance)")
print("dist(v1, v2) =", dist_v1_v2)
print("→ 크기는 다르지만 방향이 같아도 거리는 큼")
print()
print("dist(v1, v3) =", dist_v1_v3)
print("→ 공간상에서 실제로 얼마나 떨어져 있는지")
print()

# -----------------------------
# 4. 비교 요약
# -----------------------------
print("③ 코사인 vs 유클리드 비교")
print("v1 vs v2:")
print("  코사인 유사도 :", cos_v1_v2)
print("  유클리드 거리 :", dist_v1_v2)
print()
print("v1 vs v3:")
print("  코사인 유사도 :", cos_v1_v3)
print("  유클리드 거리 :", dist_v1_v3)

# -----------------------------
# 5. 정규화 후 유클리드 거리 (참고)
# -----------------------------
v1_unit = v1 / np.linalg.norm(v1)
v2_unit = v2 / np.linalg.norm(v2)

norm_dist = euclidean_distance(v1_unit, v2_unit)

print()
print("④ 정규화 후 유클리드 거리")
print("dist(unit(v1), unit(v2)) =", norm_dist)
print("→ 방향만 비교 → 코사인 유사도와 동일한 정보")
```

```python
# 코사인 유사도 vs 유클리드 거리
# 그림 + 애니메이션으로 직관적으로 보여주는 코드 (재실행)

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display

# 기준 벡터 a (고정, 단위벡터)
a = np.array([1.0, 0.0])

# 회전 각도 (0~180도)
angles = np.linspace(0, np.pi, 120)

fig, ax = plt.subplots(figsize=(6, 6))
ax.set_aspect("equal")
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

# 좌표축
ax.axhline(0)
ax.axvline(0)

# 고정 벡터 a
a_line, = ax.plot([0, a[0]], [0, a[1]], linewidth=3)
ax.text(1.05, 0.05, "a")

# 움직이는 벡터 b
b_line, = ax.plot([], [], linewidth=3)
b_label = ax.text(0, 0, "b")

# 거리 선 (a 끝점 ↔ b 끝점)
dist_line, = ax.plot([], [], linestyle="--")

title = ax.set_title("")

def update(i):
    theta = angles[i]
    b = np.array([np.cos(theta), np.sin(theta)])  # 단위벡터

    # 벡터 b 갱신
    b_line.set_data([0, b[0]], [0, b[1]])
    b_label.set_position((b[0]*1.1, b[1]*1.1))

    # 유클리드 거리 시각화
    dist_line.set_data([a[0], b[0]], [a[1], b[1]])

    # 수치 계산
    cos_sim = np.dot(a, b)
    euclid_dist = np.linalg.norm(a - b)

    title.set_text(
        f"cosine similarity = cos(θ) = {cos_sim:.2f}\n"
        f"euclidean distance = ||a - b|| = {euclid_dist:.2f}"
    )

    return b_line, dist_line, title

anim = FuncAnimation(fig, update, frames=len(angles), interval=80, blit=False)

display(HTML(anim.to_jshtml()))
```
