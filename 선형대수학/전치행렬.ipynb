전치행렬(Transpose Matrix)은 행렬의 행(row)과 열(column)을 서로 바꾼 행렬이다.
즉, 행 → 열 ,  열 → 행으로 뒤바뀐 행렬을 의미한다.

### 두 번 전치하면 원래 행렬로 돌아옴

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

print((A.T).T)
```

### 합의 전치 = 전치의 합

```python
B = np.array([[5, 6],
              [7, 8]])

left = (A + B).T
right = A.T + B.T

print(left)
print(right)
```

### 곱의 전치 = 순서를 바꾼 전치의 곱

```python
left = (A @ B).T
right = B.T @ A.T

print(left)
print(right)
```

### 스칼라 배의 전치

```python
c = 3
left = (c * A).T
right = c * A.T

print(left)
print(right)
```

### 내적(dot product) 표현

```python
import numpy as np

x = np.array([1, 3, -2])
y = np.array([4, -1, 5])

dot1 = np.dot(x, y)
dot2 = x.T @ y   # @는 행렬곱 기호

print(dot1, dot2)
```
